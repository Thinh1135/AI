```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
windspeed=ctrl.Antecedent(np.arange(0.7,3.2,0.1),'windspeed')
rainfall=ctrl.Antecedent(np.arange(0.2,9.2,0.1),'rainfall')
temperature=ctrl.Consequent(np.arange(20.4,41.3,0.1),'teamerature') 
windspeed['very low']=fuzz.trimf(windspeed.universe,[0.7,0.7,1.3])
windspeed['low']=fuzz.trimf(windspeed.universe,[0.7,1.3,1.9])
windspeed['normal']=fuzz.trimf(windspeed.universe,[1.3,1.9,2.5])
windspeed['high']=fuzz.trimf(windspeed.universe,[1.9,2.5,3.1])
windspeed['very high']=fuzz.trimf(windspeed.universe,[2.5,3.1,3.1])


temperature['very low']=fuzz.trimf(temperature.universe,[20.4,20.4,25.6])
temperature['low']=fuzz.trimf(temperature.universe,[20.4,25.6,30.8])
temperature['normal']=fuzz.trimf(temperature.universe,[25.6,30.8,36])
temperature['high']=fuzz.trimf(temperature.universe,[30.8,36,41.2])
temperature['very high']=fuzz.trimf(temperature.universe,[36,41.2,41.2])


rainfall['vl']=fuzz.trimf(rainfall.universe,[0.2,0.2,2.4])
rainfall['l']=fuzz.trimf(rainfall.universe, [0.2,2.4,4.6])
rainfall['n']=fuzz.trimf(rainfall.universe,[2.4,4.6,6.8])
rainfall['h']=fuzz.trimf(rainfall.universe, [4.6,6.8,9.1])
rainfall['vh']=fuzz.trimf(rainfall.universe, [6.8,9.1,9.1])
windspeed.view()
temperature.view()
rainfall.view()


```

```python
rule1=ctrl.Rule(windspeed['very low'] & temperature['very low'],rainfall ['vl'])
rule2=ctrl.Rule(windspeed['very low'] & temperature['low'],rainfall ['vl'])
rule3=ctrl.Rule(windspeed['very low'] & temperature['normal'],rainfall ['l'])
rule4=ctrl.Rule(windspeed['very low'] & temperature['high'],rainfall ['l'])
rule5=ctrl.Rule(windspeed['very low'] & temperature['very high'],rainfall ['n'])
rule6=ctrl.Rule(windspeed['low'] & temperature['very low'],rainfall ['vl'])
rule7=ctrl.Rule(windspeed['low'] & temperature['low'],rainfall ['vl'])
rule8=ctrl.Rule(windspeed['low'] & temperature['normal'],rainfall ['n'])
rule9=ctrl.Rule(windspeed['low'] & temperature['high'],rainfall ['h'])
rule10=ctrl.Rule(windspeed['low'] & temperature['very high'],rainfall ['h'])
rule11=ctrl.Rule(windspeed['normal'] & temperature['very low'],rainfall ['l'])
rule12=ctrl.Rule(windspeed['normal'] & temperature['low'],rainfall ['l'])
rule13=ctrl.Rule(windspeed['normal'] & temperature['normal'],rainfall ['n'])
rule14=ctrl.Rule(windspeed['normal'] & temperature['high'],rainfall ['h'])
rule15=ctrl.Rule(windspeed['normal'] & temperature['very high'],rainfall ['h'])
rule16=ctrl.Rule(windspeed['high'] & temperature['very low'],rainfall ['l'])
rule17=ctrl.Rule(windspeed['high'] & temperature['low'],rainfall ['n'])
rule18=ctrl.Rule(windspeed['high'] & temperature['normal'],rainfall ['h'])
rule19=ctrl.Rule(windspeed['high'] & temperature['high'],rainfall ['vh'])
rule20=ctrl.Rule(windspeed['high'] & temperature['very high'],rainfall ['vh'])
rule21=ctrl.Rule(windspeed['very high'] & temperature['very low'],rainfall ['n'])
rule22=ctrl.Rule(windspeed['very high'] & temperature['low'],rainfall ['h'])
rule23=ctrl.Rule(windspeed['very high'] & temperature['normal'],rainfall ['h'])
rule24=ctrl.Rule(windspeed['very high'] & temperature['high'],rainfall ['vh'])
rule25=ctrl.Rule(windspeed['very high'] & temperature['very high'],rainfall ['vh'])

Rain_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24,rule25])

Rainl=ctrl.ControlSystemSimulation(Rain_ctrl)
Rain.input['windspeed']=2.1
Rain.input['temperature']=22
Rain.compute()
print(Rain.output['rainfall'])
rainfall.view(sim=Rain)
```
